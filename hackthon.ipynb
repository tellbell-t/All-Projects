```python
Olympic_Data = pd.read_csv(io.BytesIO(uploaded['olympic_medals.csv']))
display(Olympic_Data)
```

```python
# Remove event title, event gender
```

```python
Olympic_DC=Olympic_Data.drop(columns=['event_title','event_gender','participant_type','athlete_url','athlete_full_name','country_code','participant_title','country_name'],axis=1,)
display(Olympic_DC)
```

```python
# Visualisations: 1) The Past Team Gold Medal in the summer games
```

```python
summer_sport=Olympic_DC.loc[Olympic_DC['discipline_title']=='Weightlifting']
print(summer_sport)
```

```python
# Visualition of Total Medals from Each Country
```

```python
import plotly.express as px
```

```python
medal_counts=Olympic_Data.groupby(["country_3_letter_code","medal_type"]).size().reset_index(name="count")
fig = px.bar(medal_counts, x="country_3_letter_code", y="count", color= "medal_type", title="Total Medals from Each Country")
fig.show()
```
